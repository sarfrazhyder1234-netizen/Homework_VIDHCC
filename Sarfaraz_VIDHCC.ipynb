Create a House class that represents a house; attributes:


*   area
*   number of rooms
*   number of bathrooms
*   number of floors
*   has garage
*   has garden
*   any other attributes that you think is important



instance methods:

*   price calculation

There should be at least 1 default parameter

(4 points)

```python
class House:
    def __init__(self, area, rooms, bathrooms, floors, has_garage=False, has_garden=False):
        self.area = area
        self.rooms = rooms
        self.bathrooms = bathrooms
        self.floors = floors
        self.has_garage = has_garage
        self.has_garden = has_garden


    def calculate_price(self, price_per_sqm=3000):  # default parameter

        price = self.area * price_per_sqm

        if self.has_garage:
            price *= 1.15          # +15% if garage

        if self.has_garden:
            price *= 1.06          # +6% if garden

        return price
```

```python
Myhouse = House(240, 6, 2, 1)
print(Myhouse.calculate_price())
```

Create a Person class with the following attributes:


*   name
*   age
*   married (T/F)
*   number of children
*   number of cars
*   total amount of money (to buy a house)
*   based on the number of children store the number of family members in an attribute  (- this does not have to be included in the init arguments as it is a calculated value)


(4 points)




```python
class Person:
    def __init__(self, name, age, children, cars, money, married=False):
        self.name = name
        self.age = age
        self.married = married
        self.children = children
        self.cars = cars
        self.money = money

        # Total members
        if self.married:
            self.family_members = self.children + 2
        else:
            self.family_members = self.children + 1
```

```python
Ali = Person("Ali", 26, 3, 1, 10000, True)

print("Family members:", Ali.family_members)
```

 Create an Animal class and choose 3 animals that have some things in common (characteristics and behaviour)!

Write the Parent class and the derived classes of the species!

The Child classes should inherit some attributes and override some methods of the Parent class, but should also extend the state with own attributes in each subclass and extend the behaviour with own methods.

Decide yourself:

which should be overridden


which attributes are shared. Explain your choice.

```python
class Animal:
    def __init__(self, name, age, weight):
        self.name = name
        self.age = age
        self.weight = weight

    def eat(self):
        print(f"{self.name} is eating")

    def sleep(self):
        print(f"{self.name} is sleeping")

    def make_sound(self): #will be used as overriden method in other child classes
        print("Animal sound")
```

```python
class Dog(Animal):
    def __init__(self, name, age, weight, breed):
        super().__init__(name, age, weight)
        self.breed = breed

    def make_sound(self):
        print(f"{self.name} says Woof!")

    def fetch_ball(self):   #specific attribute
        print(f"{self.name} is fetching the ball")


class Eagle(Animal):
    def __init__(self, name, age, weight, colour):
        super().__init__(name, age, weight)
        self.colour = colour

    def make_sound(self):
        print(f"{self.name} screeches")

    def hunt(self):     #specific attribute
        print(f"{self.name} is hunting prey")


class Shark(Animal):
    def __init__(self, name, age, weight, length):
        super().__init__(name, age, weight)
        self.length = length

    def make_sound(self):
        print(f"{self.name} cannot make loud sounds")

    def swim(self):       #specific attribute
        print(f"{self.name} is swimming fast")

```

```python
dog = Dog("Tommy", 4, 16, "poodle")
eagle = Eagle("Martial", 2, 5, "Black")
shark = Shark("Fin", 10, 100, 12)

dog.make_sound()
dog.eat()

eagle.hunt()
eagle.sleep()

shark.swim()
shark.sleep()
```

###B) (6 points)

Create classes and use inheritance to calculate the area of different shapes (specifically, circle and square, you can add triangle or anything else if you want).

Requirements:


1.	Create a base class called Shape that has a method named area().

This method should not have any implementation
2. Create at least two subclasses:

Circle — which takes a radius as a parameter

Square — which takes a side as a parameter

write the area-calculating function for both (or for all)

3. Create objects of both Circle and Square, and print their areas.

Extra: Store these objects in a list and use a loop to print the area of each shape.

```python
import math

class Shape:
    def area(self):
        pass

# Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2


# Square
class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side ** 2


# Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width
```

```python
# Create objects
Circ = Circle(5)
Sq = Square(4)
Rect = Rectangle(6, 3)

# Print areas
print("Circle area:", Circ.area())
print("Square area:", Sq.area())
print("Rectangle area:", Rect.area())


# Extra
shapes = [Circ, Sq, Rect]

print("All shapes:")
for s in shapes:
    print(s.area())
```

```python
names = ["Anna", "Bob", "Charlie"]
scores = [85, 92, 78]
```

Expected output: [('Anna', 85), ('Bob', 92), ('Charlie', 78)]
